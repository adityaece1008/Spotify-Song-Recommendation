```python
# Don't edit this cell, this create your playlist
import random,string
songids=['SONGID'+str(i).zfill(2) for i in range(1,21)]
userid=[''.join('USER')+str(random.randint(10,99)).zfill(2) for i in range(100)]
playlist={i:j for i,j in zip(userid,[random.sample(songids,random.randint(4,7)) for i in range(len(userid))])}
playlist
```

Let's say the new user has listened to these 3 songs

```python
myplist=['SONGID01','SONGID06','SONGID16']
```

What songs would you recommend him.

```python
recommended_user=[]
recommended_songs=[]
final_recommendation=[]
newlist=[]


for song in myplist:             #finding the user who is listening the myplist song
    for user in playlist:
        for songs in playlist[user]:
            if song==songs:
                recommended_user.append(user)
                
for user in recommended_user:          # finding the songs of user who is listening the myplist song ... 
    for song in playlist[user]:          #...means song prefer by users
        recommended_songs.append(song)
        
for songid in recommended_songs:      # finding the most preferred songs by count ....
    n=recommended_songs.count(songid)
    newlist.append(n)
    
m=min(newlist)
p=max(newlist)

for songid in recommended_songs:
    n=recommended_songs.count(songid)
    if n>=p-10 or n>=p-m or n>m+10:                          # To find the...... top most prominent song
        if songid not in myplist and songid not in final_recommendation:
            final_recommendation.append(songid)
           

        
print(final_recommendation)        
 
 
   
        
                

    
    
```
